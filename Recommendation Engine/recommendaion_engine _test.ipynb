```python

import pandas as pd

def remove_outliers_iqr(df, column):
   
    Q1 = df[column].quantile(0.25)  # First Quartile
    Q3 = df[column].quantile(0.75)  # Third Quartile
    IQR = Q3 - Q1  # Interquartile Range
    
    # Define the bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Filter the DataFrame to remove outliers
    df_filtered = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    
    return df_filtered

# Remove outliers from 'TransactionAmount'
transaction_data = remove_outliers_iqr(transaction_data, 'TransactionAmount')

# Remove outliers from 'DiscountAmount'
transaction_data = remove_outliers_iqr(transaction_data, 'DiscountAmount')



```

## Collaborative Filtering on Interaction Data
1.1 Import Required Libraries

```python
from sklearn.metrics.pairwise import cosine_similarity
import scipy.sparse as sp
```

```python
interaction_Data.head()
```

```python
weights = {
    'ViewedPages': 2,
    'ProductViewHistory': 3,
    'SearchQueries': 5,        
    'TimeSpentOnPages': 1,     
    'WishlistItems': 7,
    'CartAddedItems': 9,       
    'CartAbandonedItems': 1,
    'ReviewRating': 7
}

# Function to compute composite score
def calculate_composite_score(row):
    score = 0
    for feature, weight in weights.items():
        score += row[feature] * weight
    return score
```

1.2 Prepare Interaction Matrix

```python
# Add a composite score to the interaction data
interaction_Data['CompositeScore'] = interaction_Data.apply(calculate_composite_score, axis=1)

# Create the interaction matrix: Users x Products with composite scores
interaction_matrix = interaction_Data.pivot_table(
    index='UserID', columns='ProductID', values='CompositeScore', fill_value=0
)

```

```python
interaction_matrix.head()
```

1.3 User-Based Collaborative Filtering

```python
# Compute User Similarity Matrix
user_similarity = cosine_similarity(interaction_matrix)
user_similarity_df = pd.DataFrame(user_similarity, index=interaction_matrix.index, columns=interaction_matrix.index)
def user_based_cf_recommendations(user_id, top_n=31):
    similar_users = user_similarity_df[user_id].sort_values(ascending=False)[1:top_n+1]
    similar_user_ids = similar_users.index

    recommendations = interaction_matrix.loc[similar_user_ids].sum(axis=0)
    recommendations = recommendations[recommendations > 0].sort_values(ascending=False)
    return recommendations.index[:top_n].tolist()

# Example: Recommend for a specific user
print(user_based_cf_recommendations("0000d51a-e6b0-4333-8007-a3824b1d53a2", top_n=5))
```

```python
# Function to fetch product details for a list of ProductIDs
def fetch_product_details(product_ids):
    # Filter the product_data DataFrame to get details for the given products
    return product_data[product_data['ProductID'].isin(product_ids)][
        ['ProductID', 'ProductName', 'Category', 'BasePrice', 'Description']
    ]
```

```python
# Example: Recommend for a specific user and fetch product details
user_id = "0000d51a-e6b0-4333-8007-a3824b1d53a2"
recommended_products = user_based_cf_recommendations(user_id)
product_details = fetch_product_details(recommended_products)
product_details
```

1.4 Item-Based Collaborative Filtering

```python
# Compute Item Similarity Matrix
item_similarity = cosine_similarity(interaction_matrix.T)
item_similarity_df = pd.DataFrame(item_similarity, index=interaction_matrix.columns, columns=interaction_matrix.columns)

# Recommend Similar Products
def item_based_cf_recommendations(product_id, top_n=5):
    similar_items = item_similarity_df[product_id].sort_values(ascending=False)[1:top_n+1]
    return similar_items.index.tolist()


```

## 2. Content-Based Filtering on Product Features
2.1 Compute Product Similarity

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler

# Combine numeric and textual features
vectorizer = TfidfVectorizer(max_features=100)
product_description_vectors = vectorizer.fit_transform(product_data['Description'])

scaler = StandardScaler()
product_numeric_features = scaler.fit_transform(product_data[['BasePrice', 'StockQuantity', 'DiscountAmount']])

# Concatenate features
from scipy.sparse import hstack
product_features = hstack([product_description_vectors, product_numeric_features])

# Compute Product Similarity
product_similarity = cosine_similarity(product_features)
product_similarity_df = pd.DataFrame(product_similarity, index=product_data['ProductID'], columns=product_data['ProductID'])
```

#### 2.2 Recommend Products Based on Similarity

```python
def content_based_recommendations(product_id, top_n=5):
    similar_products = product_similarity_df[product_id].sort_values(ascending=False)[1:top_n+1]
    return similar_products.index.tolist()

```

## 3. Matrix Factorization and Clustering on User Data
3.1 Clustering Users

```python
from sklearn.cluster import KMeans
# Features for clustering
clustering_features = user_data[['Gender', 'Age', 'TotalSpent', 'TotalTransactions']].fillna(0)

# Fit KMeans model
kmeans = KMeans(n_clusters=3, random_state=42)
user_data['Cluster'] = kmeans.fit_predict(clustering_features)

# Recommend products based on cluster preferences
def cluster_based_recommendations(user_id, top_n=5):
    user_cluster = user_data.loc[user_data['UserID'] == user_id, 'Cluster'].values[0]
    cluster_users = user_data[user_data['Cluster'] == user_cluster]['UserID']
    cluster_transactions = transaction_data[transaction_data['UserID'].isin(cluster_users)]
    
    # Get most purchased products in the cluster
    recommendations = cluster_transactions['ProductID'].value_counts().head(top_n).index.tolist()
    return recommendations

```

## 4. Combine Recommendations
We’ll combine recommendations by weighting each dataset-specific recommendation strategy.

```python
from collections import defaultdict

def combine_recommendations(user_id, product_id=None, weights=[0.4, 0.25, 0.15, 0.2], top_n=5):
    """
    Combines recommendations from multiple models with given weights and returns the top N recommendations.

    Args:
        user_id (str): The ID of the user for user-based recommendations.
        product_id (str, optional): The ID of the product for item/content-based recommendations. Defaults to None.
        weights (list): List of weights corresponding to different recommendation models. Defaults to [0.2, 0.6, 0.5, 0.3].
        top_n (int): Number of top recommendations to return. Defaults to 5.

    Returns:
        list: Top N recommended product IDs.
    """
    # Fetch recommendations from different models
    user_cf = user_based_cf_recommendations(user_id, top_n)
    item_cf = item_based_cf_recommendations(product_id, top_n) if product_id else []
    content_recs = content_based_recommendations(product_id, top_n) if product_id else []
    cluster_recs = cluster_based_recommendations(user_id, top_n)

    # Aggregate scores using defaultdict
    recommendation_scores = defaultdict(float)

    # Update recommendation scores from each model
    def update_scores(recs, weight):
        for product in recs:
            recommendation_scores[product] += weight

    update_scores(user_cf, weights[0])
    update_scores(item_cf, weights[1])
    update_scores(content_recs, weights[2])
    update_scores(cluster_recs, weights[3])

    # Sort recommendations by scores in descending order and retrieve top N
    sorted_recommendations = sorted(recommendation_scores.items(), key=lambda x: x[1], reverse=True)
    return [product for product, score in sorted_recommendations[:top_n]]


def display_recommendations(user_id, product_id, weights, top_n):
    """
    Fetches, displays, and returns product details for combined recommendations.

    Args:
        user_id (str): The ID of the user.
        product_id (str): The ID of the given product.
        weights (list): List of weights for recommendation models.
        top_n (int): Number of top recommendations to display.

    Returns:
        DataFrame: Details of recommended products.
    """
    # Get combined recommendations
    recommended_products = combine_recommendations(user_id, product_id, weights, top_n)

    # Fetch and display product details
    print("Given Product Details:")
    given_product_details = fetch_product_details([product_id])
    print(given_product_details)

    print("\nRecommended Product Details:")
    recommended_product_details = fetch_product_details(recommended_products)
    print(recommended_product_details)

    return recommended_product_details


# Example usage
user_id = "0001582d-b85d-4803-8167-f44bc0d70f67"
product_id = "2086fa1b-7400-40a9-b879-69ee50bb9de6"
weights = [0.4, 0.25, 0.15, 0.2]
top_n = 5

recommended_product_details = display_recommendations(user_id, product_id, weights, top_n)
```

## 2. Ranking Aggregation
Here, we combine rankings from different models. Each model assigns ranks, and we aggregate them using methods like Borda Count or Reciprocal Rank Fusion (RRF).

```python
def rank_aggregation_hybrid_recommendations(user_id, product_id=None, k=60, top_n=5):
    # Get recommendations from different models
    user_cf = user_based_cf_recommendations(user_id, k)
    item_cf = item_based_cf_recommendations(product_id, k) if product_id else []
    cluster_recs = cluster_based_recommendations(user_id, k)

    # Combine rankings using RRF
    ranking_scores = {}
    models = [user_cf, item_cf, cluster_recs]
    for model in models:
        for rank, product_id in enumerate(model):
            if product_id in ranking_scores:
                ranking_scores[product_id] += 1 / (rank + 1)
            else:
                ranking_scores[product_id] = 1 / (rank + 1)

    # Sort and return top recommendations
    sorted_recommendations = sorted(ranking_scores.items(), key=lambda x: x[1], reverse=True)
    return [product for product, score in sorted_recommendations[:top_n]]
```

```python
# Example: User and Product Details
user_id="0001582d-b85d-4803-8167-f44bc0d70f67"
product_id="2086fa1b-7400-40a9-b879-69ee50bb9de6"

# Display given product details
given_product_details = fetch_product_details([product_id])
print("Given Product Details:")
print(given_product_details)

# Get recommendations from the hybrid rank aggregation model
recommended_products = rank_aggregation_hybrid_recommendations(user_id, product_id, k=60, top_n=5)

# Display details of recommended products
recommended_product_details = fetch_product_details(recommended_products)
print("\nRecommended Products:")
recommended_product_details
```

```python
import numpy as np
from sklearn.metrics import precision_score, recall_score, f1_score, mean_squared_error

# Evaluation Metrics Functions

def calculate_precision_recall_f1(true_products, recommended_products):
    """
    Calculate Precision, Recall, and F1-Score for recommendations.
    """
    y_true = [1 if product in true_products else 0 for product in recommended_products]
    y_pred = [1] * len(recommended_products)  # Recommended products are considered positive.
    
    precision = precision_score(y_true, y_pred, zero_division=0)
    recall = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)
    
    return precision, recall, f1


def mean_average_precision(true_products, recommended_products):
    """
    Compute Mean Average Precision (MAP).
    """
    hits = 0
    avg_precision = 0.0
    for i, product in enumerate(recommended_products):
        if product in true_products:
            hits += 1
            avg_precision += hits / (i + 1)
    return avg_precision / len(true_products) if true_products else 0


def mean_reciprocal_rank(true_products, recommended_products):
    """
    Compute Mean Reciprocal Rank (MRR).
    """
    for i, product in enumerate(recommended_products):
        if product in true_products:
            return 1 / (i + 1)
    return 0


# Evaluation Function
def evaluate_recommendations(test_data, top_n=5):
    """
    Evaluate the `rank_aggregation_hybrid_recommendations` function.
    :param test_data: DataFrame with 'UserID' and 'ProductID' columns.
    :param top_n: Number of recommendations to generate for each user.
    :return: Dictionary of evaluation metrics.
    """
    precision_list, recall_list, f1_list = [], [], []
    map_list, mrr_list = [], []

    # Iterate through unique users in the test set
    for user_id in test_data['UserID'].unique():
        # True products for the user
        true_products = test_data[test_data['UserID'] == user_id]['ProductID'].tolist()

        # Get recommended products from the hybrid function
        recommended_products = rank_aggregation_hybrid_recommendations(user_id, top_n=top_n)

        # Compute metrics
        precision, recall, f1 = calculate_precision_recall_f1(true_products, recommended_products)
        map_score = mean_average_precision(true_products, recommended_products)
        mrr_score = mean_reciprocal_rank(true_products, recommended_products)

        # Append scores
        precision_list.append(precision)
        recall_list.append(recall)
        f1_list.append(f1)
        map_list.append(map_score)
        mrr_list.append(mrr_score)

    # Aggregate results
    return {
        "Precision": np.mean(precision_list),
        "Recall": np.mean(recall_list),
        "F1-Score": np.mean(f1_list),
        "MAP": np.mean(map_list),
        "MRR": np.mean(mrr_list)
    }


# Example Usage
# Assuming `test_data` is a DataFrame with 'UserID' and 'ProductID'
results = evaluate_recommendations(test_data, top_n=5)

# Print the evaluation results
print("Evaluation Metrics:")
for metric, value in results.items():
    print(f"{metric}: {value:.4f}")
```
